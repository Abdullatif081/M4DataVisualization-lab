# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import sqlite3
import matplotlib.pyplot as plt

# the read_sql_query runs the sql query and returns the data as a dataframe
conn = sqlite3.connect("m4_survey_data.sqlite")
QUERY = """
SELECT ConvertedComp
FROM master
"""
# Run the query and load the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)
# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(df['ConvertedComp'].dropna(), bins=30, edgecolor='black')
plt.title('Histogram of ConvertedComp')
plt.xlabel('ConvertedComp')
plt.ylabel('Frequency')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT Age
FROM master
"""
# Run the query and load the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)
# Plot the box plot
plt.figure(figsize=(10, 6))
plt.boxplot(df['Age'].dropna(), vert=False)
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""
# Run the query and load the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Plot the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['WorkWeekHrs'], alpha=0.5)
plt.title('Scatter Plot of Age vs. WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.grid(True)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""
# Run the query and load the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)
# Drop rows with missing values in any of the relevant columns
df = df.dropna(subset=['WorkWeekHrs', 'CodeRevHrs', 'Age'])
# Plot the bubble plot
plt.figure(figsize=(12, 8))
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=df['Age']*10, alpha=0.5, edgecolors='w', cmap='viridis')
plt.title('Bubble Plot of WorkWeekHrs vs. CodeRevHrs with Age as Bubble Size')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.colorbar(label='Age')  # Show a colorbar with age information
plt.grid(True)
plt.show()

```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
# Query to get the relevant columns and filter by age
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
"""
# Run the query and load the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Drop rows with missing values in any of the relevant columns
df = df.dropna(subset=['WorkWeekHrs', 'CodeRevHrs'])

# Compute the median values
median_values = df[['WorkWeekHrs', 'CodeRevHrs']].median()
# Prepare data for plotting
labels = ['Median WorkWeekHrs', 'Median CodeRevHrs']
values = [median_values['WorkWeekHrs'], median_values['CodeRevHrs']]

# Create the stacked bar chart
plt.figure(figsize=(8, 6))
plt.bar(labels, values, color=['skyblue', 'lightgreen'], label=['WorkWeekHrs', 'CodeRevHrs'])
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.ylabel('Median Hours')
plt.legend()
plt.show()

```
